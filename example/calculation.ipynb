```python
import cinrad
```

**Some calculations of data**

```python
f = cinrad.io.CinradReader('Z_RADR_I_Z9200_20190421190600_O_DOR_SA_CAP.BIN.BZ2')
```

Get list of reflectivity data in entire scan

```python
rl = list(f.iter_tilt(230, 'REF'))
```

Calculate vertically integrated liquid (VIL)

```python
vil = cinrad.calc.quick_vil(rl)
```

```python
print(vil)
```

Visualize it!

Note: Due to some incompatibilities, the figure is not plotted correctly as it should be, to have a look at correctly plotted figures, please go to the `Gallery` section in README.

```python
fig = cinrad.visualize.PPI(vil, dpi=50)
```

Calculate echo tops (ET)

```python
et = cinrad.calc.quick_et(rl)
```

```python
fig = cinrad.visualize.PPI(et, dpi=50)
```

Calculate composite reflectivity (CR)

```python
cr = cinrad.calc.quick_cr(rl)
```

```python
fig = cinrad.visualize.PPI(cr, dpi=50)
```
